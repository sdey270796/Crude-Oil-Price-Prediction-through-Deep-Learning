### Decomposition of the Prices

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(brent_crude['Close'], model='additive', period=365)

plt.figure(figsize=(12, 8))
plt.title('Decomposition of Brent Crude Oil Prices')

plt.subplot(411)
plt.plot(result.trend, label='Trend')
plt.legend(loc='best')
plt.grid(True)

plt.subplot(412)
plt.plot(result.seasonal, label='Seasonality')
plt.legend(loc='best')
plt.grid(True)

plt.subplot(413)
plt.plot(result.resid, label='Residuals')
plt.legend(loc='best')
plt.grid(True)

plt.subplot(414)
plt.plot(brent_crude['Close'], label='Original')
plt.legend(loc='best')
plt.tight_layout()
plt.grid(True)

plt.show()
```

#### Interpretation of the Plot

The classical decomposition of the time series represents its key components: trend, seasonality, and residuals. Here's a breakdown of what each subplot is showing:

1. **Trend Component (Top Plot):**
   - This plot shows the long-term progression of the time series data, filtering out any short-term fluctuations or seasonal effects.
   - The trend usually represents the underlying direction in the data, either upwards, downwards, or constant over time. In this case, you can see a U-shaped trend where the series initially decreases and then increases over time.

2. **Seasonality Component (Second Plot):**
   - The seasonality plot captures repetitive patterns at regular intervals. This component is what remains after the trend has been removed, and it represents cyclical patterns that repeat over specific periods (like daily, weekly, or yearly cycles).
   - In your plot, there is a strong repeating cycle, with peaks and troughs at consistent intervals.

3. **Residuals Component (Third Plot):**
   - The residuals are whats left after both the trend and seasonality components have been removed. This plot represents the random noise or irregularities in the time series that aren't explained by either the trend or the seasonality.
   - The residuals here seem to have no clear pattern, which is typically expected for random noise, although there are a few noticeable spikes indicating anomalies or outliers in the data.

4. **Original Time Series (Bottom Plot):**
   - The bottom plot shows the original time series data before decomposition. This is the raw data that contains all of the components (trend, seasonality, and noise) together.
   - You can visually compare this with the trend, seasonality, and residuals plots to see how the original series has been broken down.

In summary, this decomposition helps in understanding the underlying patterns in the data, making it easier to model or forecast the time series, especially if certain components can be isolated and modeled independently.

***Purpose of Decomposition***

This decomposition helps isolate different aspects of the data:
- **Trend** gives an insight into the long-term movement.
- **Seasonality** shows any periodic behavior.
- **Residuals** help in identifying randomness or unmodeled behavior.
- By understanding these components separately, analysts can better model, forecast, or identify underlying patterns in the time series data.

## Train-Test Split

```python
split_index = int(0.7 * len(brent_crude))

train_df = brent_crude['Close'].iloc[:split_index]
test_df = brent_crude['Close'].iloc[split_index:]

print("Training set shape:", train_df.shape)
print("Testing set shape:", test_df.shape)
```

```python
train_df.head()
```

```python
test_df.head()
```

## Scaling the Data

```python
scaler_1 = MinMaxScaler(feature_range=(0,1))
```

```python
train_array = scaler_1.fit_transform(np.array(train_df).reshape(-1,1))
train_array[:5]
```

```python
test_array = scaler_1.transform(np.array(test_df).reshape(-1,1))
test_array[:5]
```

```python
x_train_1 = []
y_train_1 = []

for i in range(100, train_array.shape[0]):
    x_train_1.append(train_array[i-100: i])
    y_train_1.append(train_array[i, 0])

x_train_1, y_train_1 = np.array(x_train_1), np.array(y_train_1)
```

```python
x_train_1.shape
```

## Running the Model

```python
model_1 = Sequential()
model_1.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_train_1.shape[1], 1)))
model_1.add(Dropout(0.2))

model_1.add(LSTM(units=60, activation='relu', return_sequences=True))
model_1.add(Dropout(0.3))

model_1.add(LSTM(units=80, activation='relu', return_sequences=True))
model_1.add(Dropout(0.4))

model_1.add(LSTM(units=120, activation='relu'))
model_1.add(Dropout(0.5))

model_1.add(Dense(units=1))
```

## Testing the Model

```python
past_100_days = train_df.tail(100)
final_df = pd.concat([past_100_days, test_df], ignore_index=True)
final_df.head()
```

```python
input_data_1 = scaler_1.fit_transform(final_df.values.reshape(-1, 1))
input_data_1 = pd.DataFrame(input_data_1, columns=[final_df.name])
input_data_1.head()
```

### Calculating RMSE/Mean

```python
brent_rmse = np.sqrt(mean_squared_error(y_test_1, y_pred_1))
print(f"RMSE: {brent_rmse}")

brent_mean = np.mean(y_test_1)
print(f"Mean: {brent_mean}")

brent_normalized_rmse = brent_rmse / brent_mean
print(f"Normalized RMSE: {brent_normalized_rmse}")
```

#### Inference

1. **RMSE (Root Mean Squared Error): 4.1829**
   - **Inference**: RMSE is the standard deviation of the residuals (prediction errors). It measures the average magnitude of the prediction error. Here, the RMSE is approximately 4.18 units, indicating that, on average, the models predictions deviate from the actual values by about 4.18 units. The smaller the RMSE, the better the model is at predicting the actual values.

2. **Mean: 31.4741**
   - **Inference**: This is the mean of the observed time series data. It represents the central tendency or average value of the data over the analyzed period, which is about 31.47 units. The mean provides context for interpreting the scale of the RMSE. In this case, the RMSE is smaller relative to the mean, indicating the errors are relatively modest compared to the overall magnitude of the data.

3. **Normalized RMSE: 0.1329 (13.29%)**
   - **Inference**: The normalized RMSE (NRMSE) expresses the RMSE as a fraction of the mean (or sometimes the range) of the data, providing a scale-independent measure of error. In this case, an NRMSE of 0.1329 means that the RMSE is about 13.29% of the mean. This suggests that the model's prediction error is about 13.29% of the average data value, which can be considered relatively low depending on the specific application.
   
   - **Practical Interpretation**: This level of error (around 13%) might be acceptable in many time series forecasting contexts, indicating that the model is reasonably accurate. However, the degree of acceptability would depend on the specific domain and the tolerance for error in the particular scenario.

***Overall Inference***
- The model seems to perform fairly well, as the error (RMSE) is not large relative to the scale of the data (mean).

## Predictions for the Next Month

```python
def predict_future_prices(model, input_data, y_test, scale_factor, num_days=30):
    last_100_days = np.copy(input_data[-100:])
    predictions = []

    for _ in range(num_days):
        pred = model.predict(last_100_days.reshape(1, -1, last_100_days.shape[1]), verbose=0)
        predictions.append(pred[0, 0])
        last_100_days = np.roll(last_100_days, shift=-1, axis=0)
        last_100_days[-1] = pred

    predictions = np.array(predictions)
    predictions = predictions * scale_factor

    initial_price = predictions[0]
    final_price = predictions[-1]
    predicted_return = ((final_price - initial_price) / initial_price) * 100

    return predictions, predicted_return
```
